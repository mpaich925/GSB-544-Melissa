```python
import pandas as pd
data1 = pd.read_csv("q1data.csv")
cleandata1 = data1.dropna()
from plotnine import ggplot, geom_point, aes, geom_boxplot
(ggplot(cleandata1, aes(x = "income", y = "life_exp", color = "four_regions"))
+ geom_point()
)
```

```python
from plotnine import geom_histogram
(ggplot(cleandata1,
aes(
  x = "income",
  fill = "four_regions"
))
+ geom_histogram()
)
```

```python
from plotnine import facet_wrap
(ggplot(cleandata1,
aes(
  x = "income",
  y = "life_exp"
))
+ geom_boxplot()
+ facet_wrap("four_regions")
)
```
